```python
def df(x):
    return (2*(x-2))

def df2(x):
    return (2*(x-3))

def f(x1,x2):
    return pow(x1-2,2) + pow(x2-3,2)

def gradient_descent(curX1, curX2, rate, maxIters):
    precision = 0.0001  #This tells us when to stop the algorithm
    f1list = []

    for aIter in range(maxIters):
        prevX1=curX1
        curX1=curX1-rate*df(curX1)

        prevX2=curX2
        curX2=curX2-rate*df2(curX2)

        diff=abs(f(curX1,curX2)-f(prevX1,prevX2))

        f1list.append(f(curX1,curX2))

        if diff<=precision:
            print("\nThe local minimum occurs at (", curX1,",", curX2,")")
            break
        print("\nIteration",aIter,"\nX1 value is",curX1, "| X2 value is", curX2, "| f1 =", f(curX1,curX2))

    return f1list

```

```python
plt.plot(gradient_descent(0, 0, 0.5, 10))
```

```python
plt.plot(gradient_descent(0, 0, 0.001, 5000))
```

```python
plt.plot(gradient_descent(3, 5, 0.001, 5000))
```

```python
def df3(x1,x2):
    return (-40*pow(x2,2)+240*x2+40*x1-240)

def df4(x1,x2):
    return (80*pow(x2,3)-720*pow(x2,2)+1922*x2-80*x2*x1+240*x1-1448)

def f2(x1,x2):
    return pow(1-(x2-3),2) + 20*pow((x1+3)-pow(x2-3,2),2)


def gradient_descent2(curX1, curX2, rate, maxIters):
    precision = 0.0001             
    flist = []

    for aIter in range(maxIters):
        prevX1=curX1
        curX1=curX1-rate*df3(curX1,curX2)

        prevX2=curX2
        curX2=curX2-rate*df4(curX1,curX2)

        diff=abs(f2(curX1,curX2)-f2(prevX1,prevX2))

        flist.append(f2(curX1,curX2))
    
        if diff<=precision:
            print("\nThe local minimum occurs at (", curX1,",", curX2,")")
            break
        print("Iteration",aIter,"\nX1 =",curX1, "| X2 =", curX2, "| f2 =", f2(curX1,curX2))

    return flist
```

```python

plt.plot(gradient_descent2(0, 0, 0.5, 1000))
```

```python
plt.plot(gradient_descent2(0, 0, 0.001, 10000))
```

```python
plt.plot(gradient_descent2(0, 0, 0.0001, 100000))
```

```python
plt.plot(gradient_descent2(4, 7, 0.001, 10000))
```

Αντίθετα για πολύ μικρές τιμές του learning rate παρατηρούμε ότι το μοντέλο θέλει περισσότερο χρόνο για να βρει το σημείο που συγκλίνει
